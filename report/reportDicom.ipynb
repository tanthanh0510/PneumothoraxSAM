```python
import pydicom
```

```python
data = pydicom.dcmread("test.dcm")
```

```python
type(data)
```

```python
for element in data:
    print(f"{element.name}")
```

```python
import matplotlib.pyplot as plt

pixel_data = data.pixel_array

# Hiển thị hình ảnh
plt.imshow(pixel_data, cmap=plt.cm.bone)
plt.title('Hình ảnh DICOM')
plt.axis('off')  # Tắt trục
plt.show()
```
